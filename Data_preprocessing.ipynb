Check Outlier

```python
def check_outliers(df):
    min_value = []
    max_value = []
    outlier_range = []
    mean = []
    Features = []
    for i in df.columns:
        min_v = round(df[i].min(),3)
        max_v = round(df[i].max(),3)
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.95)
        IQR = Q3-Q1
        lower_limit = round(Q1 - (1.5*IQR),3)
        upper_limit = round(Q3 + (1.5*IQR),3)
        outlier_range.append([lower_limit,upper_limit])
        min_value.append(min_v)
        max_value.append(max_v)
        mean.append(df[i].mean())
        Features.append(i)
    
    details_df = pd.DataFrame({ 'Feature':Features,
                                'Min_Value':min_value,
                               'Max_Value':max_value,
                               'Outlier_Range':outlier_range,
                               'Mean':mean})
    
    return details_df
```

```python
energy_columns = energy[['avg_energy', 'maxtempC', 'mintempC', 'sunHour',
       'uvIndex', 'moon_illumination', 'WindGustKmph', 'cloudcover',
       'humidity', 'precipMM', 'visibility']].columns
for i in energy_columns:
    Q1 = energy[i].quantile(0.25)
    Q3 = energy[i].quantile(0.75)
    IQR = Q3-Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    energy[i] = np.where(energy[i] > upper_limit,upper_limit,
                                         np.where(energy[i] < lower_limit,lower_limit,
                                                 energy[i]))
```
